```python
def encode_units(x):
    if x<=0:
        return 0
    elif x>=1:
        return 1
    
haha= haha.applymap(encode_units)
```

```python
(haha.iloc[:,0]==1).sum()
```

```python
freq_itemsets= mlxtend.frequent_patterns.apriori(haha, min_support=0.02, use_colnames=True)
```

```python
freq_itemsets
```

```python
rules=mlxtend.frequent_patterns.association_rules(freq_itemsets,metric='lift', min_threshold=0.5)
```
